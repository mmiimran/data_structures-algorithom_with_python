```python
# Most common methods and operationns that can be applied to set data structure
# union of two set consis of all elements
x1 = {'data','structure'}
x2 = {'python','java','c','c++','data'}
print(x1 | x2)
print(x1.union(x2))
```

```python
# Union of two sets, x1 and x2.It can be done using two methods
# 1. using the "|" operator
# 2. using the union method
x1 = {'data','structure'}
x2 = {'python','java','c','data'}
x3 = x1 | x2
print(x3)
print(x1.union(x2))
```

```python
# Intersection of sets : To intersect an "&" operator and 'intersection()' method is used
x1 = {'data','structure','c'}
x2 = {'python','java','c','data'}
print(x1 & x2)
print(x1.intersection(x2))
```

```python
# The difference between sets can be obtained using '.difference()' method and substration " - "
# Which returns a set of all elements that are in x1. not in x2
x1 = {'data','structure','c'}
x2 = {'python','java','c','data'}
print(x1.difference(x2))
print(x1 - x2)
```

```python
# Symmetric difference can be obtained using .symmetric_difference(),
# while '^' returns a set of all data items that are present in either x1 or x2,but not both.
x1 = {'data','structure','c'}
x2 = {'python','java','c','data'}

print(x1.symmetric_difference(x2))
print(x1 ^ x2)
```

```python
# To test whether a set is a subset of another, use '.issubset()' and the operator '<='
print(x1.issubset(x2))
print(x1<=x2)
```

# Deque

```python
# A deque is a double-ended queue(deque) that support append and pop elements from both sides
# Deque are implemented as double-linked list
# whicch are very efficient for insertring and deleting elements in O(1) time complexity
```

```python
from collections import deque
s = deque()  #creates an empty deque
print(s)
my_queue = deque([1,2,'Name'])
print(my_queue)
my_queue.append('age')          # insert 'age' at the right end of the list
my_queue.appendleft('age1')     # insert 'age1' at the left end of the list
print(my_queue)

my_queue.pop()                  # delete the rightmost value of the list
print(my_queue)

my_queue.popleft()              # delete thelleftmost value of the list
print(my_queue)

```

<h1> Default Dictionary</h1>

```python
# The default dictionary (defaultdict) is a subclass of the built-in dictionary class(dict)
# That has the same methods and operations as that of the dictionary class,
# With the only difference begin that it never raises a keyError, as a noramal
# Dictionary would.defaultdict is a convenient way to initialize dictionaries
```

<h3> d = defaultdict(def_value)</h3>

```python
from collections import defaultdict
dd = defaultdict(int)
words = str.split('data python data data structure data python')
for word in words:
    dd[word] += 1
print(dd)    
```

<h1>UserDict</h1>

```python
# Python supports a container, UserDict,present in the collection module, that wraps the dictionary objects
# we can add customized functions to the dictionary.
# This is very useful for applications where we want to add/update/modify the functionalities of the dictionary.
# The below examaple where pussing/adding a new data element is not allowed in the dictionary
```

```python
# we can not push to this user dictionary
from collections import UserDict
class MyDict(UserDict):
    def push(self,key,value):
        raise RuntimeError("Cannot insert")
d = MyDict({'ab':1,'bc':2,'cd':3})
print(d)
d.push('b',2)
```

<h1>UserList</h1>

```python
# A UserList is a container that wraps list objects
# it can be used to extend the functionality of the  list data structure
```

```python
# We can not push  to this user list
from collections import UserList
class MyList(UserList):
    def push(self, key):
        raise RuntimeError("Can not insert in the list")
d = MyList([11,12,13])
print(d)
d.push(2)
```
